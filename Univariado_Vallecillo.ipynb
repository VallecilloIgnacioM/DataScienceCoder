## Analisis de la variable **review_date**

```python
df_date = df_dataset
df_date['review_year'] = pd.to_datetime(list(df_dataset.review_date)).year
df_date_year = df_date.groupby(['review_year']).review_year.agg('count')
```

```python
fig, ax =plt.subplots(figsize=(16,7))
ax = sns.lineplot(data=df_date_year)

ax.set_xticks(list(df_date.review_year.unique()))
ax.set(title='Cantidad de Reviews por A√±o', xlabel='A√±os', ylabel='Cantidad')
ax.title.set_fontstyle(fontstyle='italic')
ax.title.set_size(fontsize=20)
ax.title.set_color('Red')
# Seteo de los limites del grafico
ax.set_xlim(left=1998,right=2022)
plt.grid()
```

Se puede observar como la cantidad de comentarios fue creciendo paulatinamente hasta el a√±o 2006 donde alcanzo hasta mas de 600 comentarios, para luego mantenerse en un rango de 380 a 600 comentarios anuales entre el 2006 y 2017. Posterior a esto, fue creciendo sustancialmente hasta generar un pico maximo en el 2020. 

Se puede observar como en 2021 vuelve a la tendencia de las 600 reviews por a√±o. Mientras que en 2022, a pesar de estar a mas de mitad de a√±o, las reviews no llegan a 200.  

## Analisis de la Variable **review_rating**

```python
fig, ax = plt.subplots(figsize=(15,6))

sns.distplot(df_dataset['review_rating'], hist=True, label='')
ax.set_xticks(sorted(list(df_dataset['review_rating'].unique())))
ax.set(title='Distribucion de review rating', ylabel='Densidad', xlabel='Review Rating')
ax.title.set_fontstyle(fontstyle='normal')
ax.title.set_size(fontsize=20)
ax.title.set_color('Red')
plt.grid()
```
